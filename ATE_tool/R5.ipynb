```python
R5 = Requirement5("2021-08-12 00:00:00", "2021-8-15 23:59:59")
data = R5.get_yield_data()
yield_df = pd.DataFrame(data)
yield_df
```

```python
yield_df = yield_df.loc[(yield_df["First_yield(%)"] != 'N/A')& (yield_df["Final_yield(%)"] != 'N/A')]
yield_df
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(
    yield_df["First_yield(%)"].reset_index(drop=True),
    marker = 'o',
    markersize = 3,
    markeredgecolor = 'black'
)
plt.title("Average trend-line chart for First Yield",fontsize = 15)
plt.ylabel("First Yield", fontsize = 15)
plt.grid(True, linestyle='--', alpha=0.3)

# 消除下面的注释，并修改不同的值（前两个是范围，第三个是“步幅”）， 可以得到指定范围的折线变化，避免被异常值影响
# plt.ylim(0.7, 1, 0.02)
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(
    yield_df["Final_yield(%)"].reset_index(drop=True),
    marker = 'o',
    markersize = 3,
    markeredgecolor = 'black'
)
plt.title("Average trend-line chart for Final Yield",fontsize = 15)
plt.ylabel("Final Yield", fontsize = 15)
plt.grid(True, linestyle='--', alpha=0.3)

# 消除下面的注释，并修改不同的值（前两个是范围，第三个是“步幅”）， 可以得到指定范围的折线变化，避免被异常值影响
# plt.ylim(0.7, 1, 0.02)
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(
    yield_df["Final_yield(%)"].reset_index(drop=True),
    marker = 'o',
    markersize = 3,
    markeredgecolor = 'black',
    label = 'Final_yield'
)
plt.plot(
    yield_df["First_yield(%)"].reset_index(drop=True),
    marker = 'o',
    markersize = 3,
    markeredgecolor = 'black',
    label = 'First_yield'
)
plt.legend()
plt.title("Average trend-line chart for Final Yield",fontsize = 15)
plt.ylabel("Final Yield", fontsize = 15)
plt.grid(True, linestyle='--', alpha=0.3)

# 消除下面的注释，并修改不同的值（前两个是范围，第三个是“步幅”）， 可以得到指定范围的折线变化，避免被异常值影响
# plt.ylim(0.7, 1, 0.02)
```
